```python
# Read in a data which is used for testing whether the model is imported correctly

data_path = '../data/caffe2_learning/raw_mnist/test_lmdb'

def AddInput(model, batch_size, db, db_type):
    data, label = model.TensorProtosDBInput(
        [], ['data', 'label'], batch_size = batch_size,
        db=db, db_type=db_type)
    return data, label

model = model_helper.ModelHelper(name='readindata')
AddInput(model, 32, data_path, 'lmdb')

# Run net
workspace.RunNetOnce(model.param_init_net)
workspace.RunNetOnce(model.net)
# fetch data
blob = workspace.FetchBlob('data')
label = workspace.FetchBlob('label')
workspace.ResetWorkspace()
workspace.Blobs()
```

```python
# 方法1： 用minidb load模型
from caffe2.python.predictor import predictor_exporter as pe

model_path = '1-MNIST/mnist_model.minidb'

predict_net = pe.prepare_prediction_net(model_path, 'minidb') # import the net

model = model_helper.ModelHelper(name='test model', init_params=False) # create a new model
model.net.AppendNet(predict_net) # append the net into the model

workspace.CreateNet(model.net) # create the net. NOTICE in prepare_prediction_net function, we have created a net, but the name is different

workspace.FeedBlob('data', blob)
workspace.RunNet(model.net)

print(label[:3])
print(workspace.FetchBlob('softmax')[:3])
```

```python
# 方法2： 用predict_net.pb, init_net.pb load模型
from caffe2.proto import caffe2_pb2

predict_net = '1-MNIST/predict_net.pb'
init_net = '1-MNIST/init_net.pb'

model = model_helper.ModelHelper(name='test model', init_params=False)

# bring in param_init_net
init_net_proto = caffe2_pb2.NetDef() # a class of NetDef
with open(init_net, 'rb') as f: 
    init_net_proto.ParseFromString(f.read()) 
tmp_param_net = core.Net(init_net_proto)
model.param_init_net.AppendNet(tmp_param_net)
# bring in predict net
predict_net_proto = caffe2_pb2.NetDef()
with open(predict_net, 'rb') as f:
    predict_net_proto.ParseFromString(f.read())
tmp_predict_net = core.Net(predict_net_proto)
model.net.AppendNet(tmp_predict_net)

workspace.RunNetOnce(model.param_init_net)
workspace.FeedBlob('data', blob)
workspace.CreateNet(model.net)
workspace.RunNet(model.net)
print(label[:3])
print(workspace.FetchBlob('softmax')[:3])
```

```python
# continue with nets created in method 2
def splitmodel(predict_net, init_net, target, second_net_name='second net', second_init_net_name = 'second init net'):
    # get a dict: {output:input sets}
    lookup = dict()
    for op in predict_net.Proto().op:
        output = op.output[0]
        if output not in lookup:
            lookup[output] = set(op.input)
        else:
            lookup[output].update(op.input)
    # get the target
    visited = set() # what outputs we need to keep. This is used to target ops. But notice they could be external inputs
    unvisited = [target]
    external_inputs = set() # these are external inputs
    while len(unvisited) > 0:
        curr_output = unvisited.pop()
        visited.add(curr_output) # visited = external inputs + internal inputs
        try:
            curr_inputs = lookup[curr_output]
            for input in curr_inputs:
                if input not in visited:
                    unvisited.append(input)
        except KeyError as e: # internal inputs
            external_inputs.add(curr_output)
    
    # Construct net
    ## net
    second = caffe2_pb2.NetDef()
    second.name = second_net_name
    for op in predict_net.Proto().op:
        if op.output[0] in visited:
            new_op = second.op.add()
            new_op.CopyFrom(op)
    
    ## init net
    secondinit = caffe2_pb2.NetDef()
    secondinit.name = second_init_net_name
    for op in init_net.Proto().op:
        if op.output[0] in visited:
            new_op = secondinit.op.add()
            new_op.CopyFrom(op)
    # External output
    ## net
    for ins in predict_net.Proto().external_output:
        if ins in visited:
            g = second.external_output
            g.append(ins)
#     ## init net
#     for ins in init_net.Proto().external_output:
#         if ins in visited:
#             g = secondinit.external_output
#             g.append(ins)
#     g = secondinit.external_output
#     g.append(target) # set target as external output
    
    # External input
    ## net
    for ins in predict_net.Proto().external_input:
        if ins in visited:
            g = second.external_input
            g.append(ins)
#     ## init net
#     for ins in init_net.Proto().external_input:
#         if ins in visited:
#             g = secondinit.external_input
#             g.append(ins)  
    # up to now, second and secondinit are NetDef, we need to transform them into net
    return core.Net(second), core.Net(secondinit)
```

# Change last layer's activation func and train it

```python
def AddInput(model, batch_size, db, db_type):
    data, label = model.TensorProtosDBInput(
        [], ['data', 'label'], batch_size = batch_size,
        db=db, db_type=db_type)
    return data, label
```

```python
# training model
second, second_init = splitmodel(model.net, model.param_init_net, 'fc3')
train_model = model_helper.ModelHelper(name='new model', init_params = True)
## add input operator 
data_path = '../data/caffe2_learning/raw_mnist/train_lmdb'
AddInput(train_model, 100, data_path, 'lmdb')
data = train_model.Scale('data', 'data', scale=float(1./256))
# 把不需要的部分删除掉
second.Proto().external_input.remove('data')
second_init.Proto().op.remove(second_init.Proto().op[-1]) # delete data
# append net
train_model.net.AppendNet(second)
train_model.param_init_net.AppendNet(second_init)

tanh3 = brew.relu(train_model, 'fc3', 'tanh3')
stop_grad = train_model.StopGradient(tanh3, 'fc_stop_grad')

pred = brew.fc(train_model, stop_grad, 'pred', dim_in = 500, dim_out=10)
softmax = brew.softmax(train_model, pred, 'softmax')

# add accuracy
accuracy = brew.accuracy(train_model, [softmax, 'label'], 'accuracy')

# add training ops
xent = train_model.LabelCrossEntropy([softmax, 'label'], 'xent')
loss = train_model.AveragedLoss(xent, 'loss')
train_model.AddGradientOperators([loss])
optimizer.build_sgd(
    train_model, 
    base_learning_rate = 0.01,
    policy = 'step',
    stepsize = 1,
    gamma = 0.999)
```

```python
# testing model
second, second_init = splitmodel(model.net, model.param_init_net, 'fc3')
test_model = model_helper.ModelHelper(name='test model', init_params = False)
## add input operator 
data_path = '../data/caffe2_learning/raw_mnist/test_lmdb'
AddInput(test_model, 100, data_path, 'lmdb')
data = test_model.Scale('data', 'data', scale=float(1./256))
# 把不需要的部分删除掉
second.Proto().external_input.remove('data')
second_init.Proto().op.remove(second_init.Proto().op[-1]) # delete data
# append net
test_model.net.AppendNet(second)
test_model.param_init_net.AppendNet(second_init)

tanh3 = brew.relu(test_model, 'fc3', 'tanh3')
stop_grad = test_model.StopGradient(tanh3, 'fc_stop_grad')

pred = brew.fc(test_model, stop_grad, 'pred', dim_in = 500, dim_out=10)
softmax = brew.softmax(test_model, pred, 'softmax')

# add accuracy
accuracy = brew.accuracy(test_model, [softmax, 'label'], 'accuracy')

```
